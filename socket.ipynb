# 소켓(socket)

### 호스트 
네트워크에 연결된 모든 장치들을 노드라고 하는데 이런 노드들 중 네트워크 주소(ip)를 가지고 있는 애들을 호스트라고 보면된다.  
스마트폰이든 노트북이든 인터넷에 연결되어 있으면 호스트라고 보면된다.   
이 호스트들 끼리 서로 데이터를 주고 받는다. 

### 프로세스  
프로그램, 애플리케이션, 앱, 소프트웨어 => 프로세스  
한 컴퓨터나 스마트폰(호스트)는 여러개의 프로세스가 존재하며, 이 프로세스 들은 동시에 혹은 마치 동시라고 느껴지게끔 동작한다.  
네트워크 공간상에서 실제로 데이터를 주고 받는 것인 호스트안의 프로세스이다. 

### 포트  
데이터가 호스트가 아닌 프로세스에서 주고받는다 라는 것을 이해했고, 호스트에서는 여러 프로세스가 동시에 동작한다, 라는 것을 이해했다면  
데이터가 네트워크를 타고 호스트에 도착하면 끝이 아니고 정확한 프로세스에 도착해야 한다는 것을 알 수 있다. 
이를 위해 포트가 등장한다.  
포트란 네트워크를 통해 데이터를 주고 받는 프로세스를 식별하기 위해 호스트 내부적으로 프로세스가 할당받는 고유한 값이다.  

포트는 숫자로 표현되기 때문에 포트 넘버라고도 한다.   
그럼 포트는 데이터를 받기 위해서만 필요한 걸까? 아니다. 포트는 데이터를 보내는데에도 필요하다.   
만약 내가 받은 데이터에 대해 답변을 하고 싶은데 발신인이 없다면?? 더이상의 연결은 불가능 할 것이다.   
따라서 데이터를 담아 보내는 쪽의 프로세스도 반드시 포트 번호를 할당받은 상태여야한다.  

### 소켓  
간단하게 소켓은 프로세스가 열어야하는 창구 같은 것이다.  
만약 우체원이 발신인과 수신인이 제대로 적혀있는 편지를 전달하려고 하는데 수신인의 우편함의 입구가 막혀있다면? 이도저도 못하는 상황이 발생할 것이다.  
소켓을 열기 위해선 호스트에 할당된 ip, 포트 넘버, 프로토콜 등이 필요하며   
이 세가지가 소켓을 정의한다.


프로세스가 네트워크 통신을 하기 위해서는 포트를 할당받아야 하는데, 서버의 경우에는 보통 하나만 할당받는다.  
왜냐하면 같은 프로세스가 같은 포트와 프로토콜을 가지고도 여러개의 소켓을 열 수 있기 때문이다.  
흥미롭게도 하나의 프로세스는 같은 포트와 프로토콜을 가진 소켓을 수십 수만개를 열 수 있다. 게임의 동시접속자가 몇만이 가능한것도 같은 원리이다.  
소켓을 정의하는 것과 소켓을 식별하는 것을 구분하자  
ip주소, 포트넘버, 프로토콜로 소켓을 정의할 수 있지만 이 것으로 소켓을 식별할 수는 없다.  


### 소켓관련 에러
